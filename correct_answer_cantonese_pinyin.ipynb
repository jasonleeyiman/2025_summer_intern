```python
file_path='/teamspace/studios/this_studio/location_list/locations.txt'
location=[]
with open(file_path, 'r', encoding='utf-8') as file:
    for line in file:
        location.append(line.strip())
location
```

```python
import pycantonese
def chinese_to_jyuping(text):
    jyutping_list = pycantonese.characters_to_jyutping(text)
    return jyutping_list
for item in location:
    jyuping=chinese_to_jyuping(item)
    print(jyuping)
```
