```python
import torch
from transformers import AutoModelForSpeechSeq2Seq, AutoProcessor, pipeline
from datasets import load_dataset
import glob
device = "cuda:0" if torch.cuda.is_available() else "cpu"
torch_dtype = torch.float16 if torch.cuda.is_available() else torch.float32
```

```python
import torch
from transformers import AutoModelForSpeechSeq2Seq, AutoProcessor, pipeline
from datasets import load_dataset
model_id = "openai/whisper-large-v3-turbo"
model = AutoModelForSpeechSeq2Seq.from_pretrained(
    model_id, torch_dtype=torch_dtype, low_cpu_mem_usage=True, use_safetensors=True
)
model.to(device)
processor = AutoProcessor.from_pretrained(model_id)
# forced_decoder_ids = processor.get_decoder_prompt_ids(language="zh", task="transcribe", "repetition_penalty": 1.5,)
pipe = pipeline(
    "automatic-speech-recognition",
    model=model,
    tokenizer=processor.tokenizer,
    feature_extractor=processor.feature_extractor,
    torch_dtype=torch_dtype,
    device=device,
    generate_kwargs={
        # "forced_decoder_ids": forced_decoder_ids
        "language": "zh",  # 
        "task": "transcribe",
        "temperature": 0.0,  # 
        "repetition_penalty": 1.5,  # 
        # "no_repeat_ngram_size": 1,
    },
)
```

```python
predicted_result_temp=[]
for item in common_voice['audio']:
    result=pipe(item)
    print(result['text'])
    predicted_result_temp.append(result['text'])
predicted_result_temp
```

```python
def merge_words_and_pinyins(data):
    filtered_data = [(word, pinyin) for word, pinyin in data if pinyin is not None]
    # words = ''.join([word for word, pinyin in filtered_data])
    remove_digits = str.maketrans('', '', '0123456789')
    pinyins = ' '.join([pinyin for word, pinyin in filtered_data])
    return f"{pinyins.translate(remove_digits)}"
```

```python
import ToJyutping
output_pingyin=[]
for item in predicted_result_temp:
    pingyin=ToJyutping.get_jyutping_list(item)
    print(merge_words_and_pinyins(pingyin))
    output_pingyin.append(merge_words_and_pinyins(pingyin))
```

```python
import ToJyutping
formatted_correctanswer=[]
for item in common_voice['text']:
    pingyin=ToJyutping.get_jyutping_list(item)
    print(merge_words_and_pinyins(pingyin))
    formatted_correctanswer.append(merge_words_and_pinyins(pingyin))
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
text_model = SentenceTransformer('intfloat/multilingual-e5-base')
predicted_result=[]
correct_answers=[]
wrong_index=[]
index=0
for item in output_pingyin:
    transcribed_embedding = text_model.encode([item])
    choice_embeddings = text_model.encode(formatted_correctanswer)
    similarities = cosine_similarity(transcribed_embedding, choice_embeddings)[0]
    predicted_index = np.argmax(similarities)
    predicted_choice = common_voice['text'][predicted_index]
    confidence = similarities[predicted_index]
    predicted_result.append(predicted_choice)
    correct_answers.append(common_voice['text'][index])
    if predicted_choice==common_voice['text'][index]:
        print(f"Predicted: '{predicted_choice}' with {confidence:.4f} similarity, correct answer: {common_voice['text'][index]}")
    else:
        print(f"Predicted: '{predicted_choice}' with {confidence:.4f} similarity, correct answer: {common_voice['text'][index]}, not correct")
        wrong_index.append(index)
    index+=1
```
